```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
country = ['Venezuela','Guatemala','Brazil','Georgia','United States','Canada']
deaths = [49.73, 25.48, 22, 15.56, 12.21, 2]
ownership = [18.5, 12.1, 8.3, 10.1, 120.5, 34.7]

dict = {'country':country, 'deaths':deaths, 'ownership':ownership}
```

```python
df = pd.DataFrame(dict)
df = df.set_index('country')
df
```

```python
labels = country
x = ownership
y = deaths

rng = np.random.RandomState(0)                     # randomizer with numpy
colors = rng.rand(6)                               # Randomizes 6 colors for the 6 Countries

fig, ax = plt.subplots(1, figsize=(16, 10))        # Plotter 


plt.grid(color='k', linestyle='--', linewidth=0.1) # Grid color/size/linestyle

ax.scatter(x, y,              # Plot the scatter points
           cmap="viridis",    # Colormap
           c=colors,          # Color of the dots
           s=200,             # Size of the dots
           alpha=0.5,         # Alpha of the dots
           edgecolor='black', # Edge color of the dots
           linewidths=2)      # Size of edge around the dots

for x_pos, y_pos, label in zip(x, y, labels):
    
    ax.annotate(label,                      # The label for this point
                xy=(x_pos, y_pos),          # Position of the corresponding point
                xytext=(-30, 15),           # Offset text by 7 points to the right
                textcoords='offset points', # tell it to use offset points
                ha='left',                  # Horizontally aligned to the left
                va='center')                # Vertical alignment is centered
    
plt.xlabel('Guns per 100 People')
plt.ylabel('Gun Related Deaths per 100k People')
plt.title('Countries Gun Ownership and Deaths') # Main Title
plt.show()
```
